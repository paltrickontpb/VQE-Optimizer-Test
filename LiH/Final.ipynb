```python
import numpy as np
import matplotlib.pyplot as plt

# Import the 4 npy files
data1 = np.load("1-GD.npy")
data2 = np.load("2-Adam.npy")
data3 = np.load("3-NAG.npy")
data4 = np.load("4-Adagrad.npy")

steps = np.arange(0, 101)

# Plot the data on a xy graph
plt.plot(steps , data1, label="Gradient Descent")
plt.plot(steps , data2, label="Adam Optimizer")
plt.plot(steps , data3, label="NAG Optimizer")
plt.plot(steps , data4, label="Adagrad Optimizer")

# Add a legend
plt.legend()

# Set the title and labels of the graph
plt.title("Optimizer Performance Comparison")
plt.ylabel("Energy level in Hartree (Ha)")
plt.xlabel("Steps")

# Show the plot
plt.show()
```
